# Step 2 : Vectorize text into embedding vectors

We wil beusign SEntence-Transformers library to load a language model to vectorixze our textinto embeddings. The library hosts some of the most popular tranformers on Hugging Face Model Hub . here ,we ar using the model = SentenceTransformer ("all-MiniLM-L6-v2") to generate embeddings.

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2" , cache_folder = "data/")
faiss_title_embedding = model.encode(pdf_subset.title.values.tolist())
len(faiss_title_embedding),len(faiss_title_embedding[0])
```

# **Step 3: Saving embedding vectors to FAISS index**
Below, we create the FAISS index objects based on our embedding vectors, normalize vectors,and add these bectors to the FAISS index.


```python
import numpy as np
```

```python
import faiss

pdf_to_index = pdf_subset.set_index(["id"], drop=False)
id_index = np.array(pdf_to_index.id.values).flatten().astype("int")

content_encoded_normalized = faiss_title_embedding.copy()
faiss.normalize_L2(content_encoded_normalized)

# The IndexFlatIP belo builds index
index_content = faiss.IndexIDMap(faiss.IndexFlatIP(len(faiss_title_embedding[0])))
index_content.add_with_ids(content_encoded_normalized,id_index)
```

# **Step 4 : Search for relevant documents from the bulk of Document**

We define a search function below to first vectoriz our query text,and then search for the vectors with the closest distance.


```python
def search_content(query,pdf_to_index,k=3):
  query_vector = model.encode([query])
  faiss.normalize_L2(query_vector)

  # Here k is set to limit thenumber of Vectors we want to return
  # From Here code is written for Implementing Searching algorithm of Kth nearest neighbour
  top_k = index_content.search(query_vector,k)
  ids= top_k[1][0].tolist()
  similarities = top_k[0][0].tolist()
  results = pdf_to_index.loc[ids]
  results["similarities"]= similarities
  return results
```
