# **Step 2 : Vectorize text into embedding vectors**

I will be using sentence_Transformers library t load a language model tovectorize our text into embeddings.  This library hosts some of the most popular transformers on Hugging Face Model Hub . In this Notebook , I am using model = SentenceTransformers("all-MiniLM-L6-v2") to generate Embedding

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2" , cache_folder = "data/")
faiss_title_embedding = model.encode(pdf_subset.title.values.tolist())
len(faiss_title_embedding),len(faiss_title_embedding[0])
```

# **Step 3: Saving embedding vectors to FAISS index**
Below, we create the FAISS index objects based on our embedding vectors, normalize vectors,and add these bectors to the FAISS index.


```python
import numpy as np
```

```python
import faiss

pdf_to_index = pdf_subset.set_index(["id"], drop=False) # here we are Storing vector Indices in array
id_index = np.array(pdf_to_index.id.values).flatten().astype("int") # this array has all of the Vector ID's

content_encoded_normalized = faiss_title_embedding.copy()
faiss.normalize_L2(content_encoded_normalized) # Here I am normalizing this Vectors so that vectors will have unique length so that their dot product will have unique cosine Similarity which wil help in searching

# The IndexFlatIP below builds index
# IndexIDMap translates search results into IDs.
# Index here is refering to file Index
index_content = faiss.IndexIDMap(faiss.IndexFlatIP(len(faiss_title_embedding[0])))
""" here we are Passing Dimension of Embedding Vector,
 here IndexIDMap and IndexFlatIP is api's that we are Invoking and in both jargon IP = Inner Product we are maximisinig Inner Product between
 Queries and Document . And "Flat" stands for that this is no Vector compressing storage is same, but there are other apis in FAISS that having capability of
 Vector compressioni"""
index_content.add_with_ids(content_encoded_normalized,id_index) # here we are mapping actual Vector with Vector IDs
```

# **Step 4 : Search for relevant documents from the bulk of Document**

We define a search function below to first vectoriz our query text,and then search for the vectors with the closest distance.


```python
# till here we Convert our document into Vectors and also Index it and similarty below function will convert query into Embedding and then it will store it into FAISS
def search_content(query,pdf_to_index,k=3):
  query_vector = model.encode([query])
  faiss.normalize_L2(query_vector)# smae Normalizing Process as Previous one

  # Here k is set to limit thenumber of Vectors we want to return
  # From Here code is written for Implementing Searching algorithm of Kth nearest neighbour
  top_k = index_content.search(query_vector,k)
  ids= top_k[1][0].tolist()
  similarities = top_k[0][0].tolist()
  results = pdf_to_index.loc[ids]
  results["similarities"]= similarities
  return results
```
